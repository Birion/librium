```python
from librium.database import db_session, Book
import requests
from bs4 import BeautifulSoup
from furl import furl
from typing import Union
from time import sleep
from pathlib import Path
```

```python
url = furl("https://www.kosmas.cz/hledej")
```

```python
books = Book.query.filter(Book.has_cover.is_(False)).all()
```

```python
def get_cover_url(isbn: int) -> Union[str, None]:
    sleep(1)
    local_url = url.copy()
    local_url.path.segments += [isbn]
    r = requests.get(local_url)
    if r.status_code != requests.codes.ok:
        return None
    img = BeautifulSoup(r.text, "html5lib").find("img", id="detailCover")
    if not img:
        return None
    return img["src"]
```

```python
def store_cover(book):
    url = get_cover_url(book.isbn)
    if url:
        print(url)
        r = requests.get(url)
        ext = Path(url).suffix
        with open(f"covers/{book.uuid}{ext}", "wb") as fp:
            fp.write(r.content)
        book.has_cover = True
```

```python
for book in books:
    print(book.title)
    store_cover(book)
```

```python
db_session.commit()
```
